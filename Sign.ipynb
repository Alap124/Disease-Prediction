```python
import tkinter as tk
from tkinter import messagebox
import mysql.connector
import subprocess

# Function to add a new user to the database
def signup():
    username = username_entry.get()
    password = password_entry.get()

    db_connection = connect_to_db()
    if db_connection:
        cursor = db_connection.cursor()
        query = "INSERT INTO users (username, password) VALUES (%s, %s)"
        cursor.execute(query, (username, password))
        db_connection.commit()
        messagebox.showinfo("Sign Up Successful", "User registered successfully!")
        cursor.close()
        db_connection.close()
        open_work_notebook()
        '''cursor.close()
        db_connection.close()'''
    else:
        messagebox.showerror("Error", "Failed to connect to the database")

# Connect to MySQL database
def connect_to_db():
    try:
        db_connection = mysql.connector.connect(
            host="localhost",
            user="nec",
            password="12345",
            database="login"
        )
        return db_connection
    except mysql.connector.Error as err:
        print("Error: {}".format(err))
        return None
# Function to open work notebook
def open_work_notebook():
    subprocess.Popen(["jupyter", "notebook", "connect.ipynb"])

# Create main window
root = tk.Tk()
root.title("Sign Up")

# Username Label and Entry
username_label = tk.Label(root, text="Username:")
username_label.pack()
username_entry = tk.Entry(root)
username_entry.pack()

# Password Label and Entry
password_label = tk.Label(root, text="Password:")
password_label.pack()
password_entry = tk.Entry(root, show="*")
password_entry.pack()

# Sign Up Button
signup_button = tk.Button(root, text="Sign Up", command=signup)
signup_button.pack()

root.mainloop()
```
